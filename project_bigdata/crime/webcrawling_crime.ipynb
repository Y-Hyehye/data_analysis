```python
from selenium import webdriver
```

```python
import pandas as pd
```

```python
driver=webdriver.Chrome('chromedriver')
```

```python
num=[]
titles=[]
registrant=[]
date=[]
people=[]

for pageno in range(1, 11):
    url='https://www.police.go.kr/user/bbs/BD_selectBbsList.do?q_bbsCode=1003&q_bbscttSn=&q_tab=&q_searchKeyTy=&q_searchVal=&q_rowPerPage=10&q_currPage=%d&q_sortName=&q_sortOrder=&' %pageno
    driver.get(url)
    driver.implicitly_wait(5)
    driver.find_element_by_css_selector("""#contents2 > div.row.ui-has-breadcrumb > div.col-md-9.contents > div > ul > li.active > a""").click()
    tags_num=driver.find_elements_by_css_selector("""#contents2 > div.row.ui-has-breadcrumb > div.col-md-9.contents > div.block.table-responsive.grid-list.for-mobile > table > tbody > tr > td.show-col.text-center""")
    tags_titles=driver.find_elements_by_css_selector("""#contents2 > div.row.ui-has-breadcrumb > div.col-md-9.contents > div.block.table-responsive.grid-list.for-mobile > table > tbody > tr > td.subject > a""")
    tags_registrant=driver.find_elements_by_css_selector("""#contents2 > div.row.ui-has-breadcrumb > div.col-md-9.contents > div.block.table-responsive.grid-list.for-mobile > table > tbody > tr > td:nth-child(4)""")
    tags_date=driver.find_elements_by_css_selector("""#contents2 > div.row.ui-has-breadcrumb > div.col-md-9.contents > div.block.table-responsive.grid-list.for-mobile > table > tbody > tr > td:nth-child(5)""")
    tags_people=driver.find_elements_by_css_selector("""#contents2 > div.row.ui-has-breadcrumb > div.col-md-9.contents > div.block.table-responsive.grid-list.for-mobile > table > tbody > tr > td:nth-child(6)""")
    
    
    
    for one in tags_num:
        num.append(one.text)
    for one in tags_titles:
        titles.append(one.text)
    for one in tags_registrant:
        registrant.append(one.text)
    for one in tags_date:
        date.append(one.text)
    for one in tags_people:
        people.append(one.text)
```
